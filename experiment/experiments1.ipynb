The goal of this dataset is to predict the price of a given diamond. Basically, it is Regression problem.

There are 10 independent variables in the data:
1) id: It is a unique identifier of each diamond.
2) carat: It refers to unique unit of the weight measurement which is used to measure or weigh gemstones or diamonds.
3) cut: It is the quality of the diamond cut.
4) color: It describes the color of the diamond.
5) clarity: Clarity measures the purity and rarity of the diamond. It is graded by the visibility of these characteristics under 10 power magnification.
6) Depth: It represents the height of the diamond.
7) Table: A diamond's table is the facet which can be seen when the diamond is viewed face up

x: Diamond X dimension 
y: Diamond Y dimension 
z: Diamond Z dimension 

Target Variable: 
price: Price of a specific diamond
